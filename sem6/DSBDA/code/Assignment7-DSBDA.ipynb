```python
documentA = 'Jupiter is the largest Planet'
documentB = 'Mars is the fourth planet from the Sun'
```

```python
documentA = documentA.lower()
documentB = documentB.lower()
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
wordsA = word_tokenize(documentA)
wordsB = word_tokenize(documentB)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words =set(stopwords.words('english'))
filteredA = [w for w in wordsA if not w in stop_words]
filteredB = [w for w in wordsB if not w in stop_words]
```

```python
uniqueWords = set(filteredA).union(set(filteredB))
```

```python
numOfWordsA = dict.fromkeys(uniqueWords, 0)
numOfWordsB = dict.fromkeys(uniqueWords, 0)
for word in filteredA:
  numOfWordsA[word] += 1
for word in filteredB:
  numOfWordsB[word] += 1
```

```python
def computeTF(wordDict, bagOfWords):
  tfDict = {}
  bagOfWordsCount = len(bagOfWords)
  for word, count in wordDict.items():
    tfDict[word] = count / float(bagOfWordsCount)
  return tfDict

tfA = computeTF(numOfWordsA, filteredA)
tfB = computeTF(numOfWordsB, filteredB)

print("\n TF for Document A :",tfA)
print("TF for Document B:",tfB)
```

```python
def conputeIDF(documents):
  import math
  N = len(documents)
  idfDict = dict.fromkeys(documents[0].keys(), 0)
  for document in documents:
    for word, val in document.items():
      if val > 0:
        idfDict[word] += 1

  for word, val in idfDict.items():
    idfDict[word] =math.log(N/ float(val))
  return idfDict
idfs=conputeIDF([numOfWordsA, numOfWordsB])
print("\n IDF values :",idfs)
```

```python
def computeTFIDF(tfBagOfWords, idfs):
  tfidf = {}
  for word, val in tfBagOfWords.items():
    tfidf[word] = val * idfs[word]
  return tfidf
tfidifA=computeTFIDF(tfA, idfs)
tfidifB=computeTFIDF(tfB, idfs)

print("\n TF-IDF for Document A :",tfidifA)
print("\n TF-IDF for Document B :",tfidifB)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
text="studies studying cries cry"
```

```python
words=word_tokenize(text)
print("Orignal Words:",words)
```

```python
ps=PorterStemmer()
stemmed_words=[]
for word in words:
  stemmed_words.append(ps.stem(word))
print("Stemmed Words:",stemmed_words)
```

```python
lematizer=WordNetLemmatizer()
leamitized_words=[]
for word in words:
  leamitized_words.append(lematizer.lemmatize(word))

print("\nLemmatized Words:",leamitized_words)
```
