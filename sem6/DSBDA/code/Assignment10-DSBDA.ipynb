```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
df_iris = pd.read_csv('/content/iris.csv')

print("First 5 rows of the Iris dataset:")
display(df_iris.head())
```

```python
print("Features and their types:")
display(df_iris.info())
```

```python
numerical_features = df_iris.select_dtypes(include=['number']).columns

plt.figure(figsize=(15, 15)) # Increased figure size to better accommodate 6 plots
for i, feature in enumerate(numerical_features):
    plt.subplot(3, 2, i + 1) # Changed to 3 rows, 2 columns to fit 5 features
    sns.histplot(df_iris[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 15)) # Increased figure size to better accommodate 6 plots
for i, feature in enumerate(numerical_features):
    plt.subplot(3, 2, i + 1) # Changed to 3 rows, 2 columns to fit 5 features
    sns.boxplot(y=df_iris[feature])
    plt.title(f'Boxplot of {feature}')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```
